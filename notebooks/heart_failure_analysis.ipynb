```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='HeartDisease', data=df)
plt.title('Heart Disease Distribution')
plt.show()
```

```python
sns.boxplot(x='HeartDisease', y='Age', data=df)
plt.title('Age vs Heart Disease')
plt.show()
```

```python
sns.boxplot(x='HeartDisease', y='MaxHR', data=df)
plt.title('Max Heart Rate vs Heart Disease')
plt.show()
```

## Key Insights

Patients with heart disease tend to have different max heart rate patterns compared to those without
Age appears to play a role in heart disease risk
Certain clinical variables may be strong indicators of heart disease presence
